## X AND Y
The data are splited into X and Y. X is the input data and Y is the output data.
20% percent of the data are use to test the data why the other 80% are use to train the data.

```python
x = new_df.drop('Class',axis = 1)
y = new_df['Class']
```

```python
# use x and y variables to split the training data into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =.20 ,random_state = 101)
```

## THE DATA HAS USE CLASSIFICATION WHICH WAS DTERMINED BY THE PREDICTED VALUS OF THE MODEL
This Data was tested on 32 classification Alghorithim and the best one was selected based on the accuracy of the model.

```python
import numpy as np
import random
```

```python

#from lazypredict.Supervised import LazyRegressor


from lazypredict.Supervised import LazyClassifier 

# Turn off notebook-specific tqdm to avoid ipywidgets error
import lazypredict.Supervised
lazypredict.Supervised.use_notebook_tqdm = False


# Initialize LazyRegressor or  LazyClassifier 
reg =  LazyClassifier (verbose=0, ignore_warnings=True)

# Fit and get results
np.random.seed(101)
random.seed(101)
models, predictions = reg.fit(x_train, x_test, y_train, y_test)

# Show model comparison
print(models)
```

## WE PICKED THE TWO BEST ALGHORITHM FOR THE TASK WHICH SHOW THE BEST RESULTS WITH AN ACCURACY OF 95%

```python
models.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=44)
# Train the model
rf_model.fit(x_train, y_train)
# Predict on the test set
y_pred1 = rf_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred1)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_pred1)
recall = recall_score(y_test, y_pred1)
f1 = f1_score(y_test, y_pred1)

print(f"Precision = {precision}")
print(f"Recall = {recall}")
print(f"F1 Score = {f1}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# Initialize the Gradient Boosting model
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=44)

# Train the model
gb_model.fit(x_train, y_train)

# Predict on the test set
y_pred2 = gb_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred2)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_pred2)
recall = recall_score(y_test, y_pred2)
f1 = f1_score(y_test, y_pred2)

print(f"Precision = {precision}")
print(f"Recall = {recall}")
print(f"F1 Score = {f1}")
```

```python
m_accuracy = pd.DataFrame({'Models':['Random Forest','Gradient Boosting'],
             "Accuracy":[accuracy_score(y_test,y_pred1)*100,
                         accuracy_score(y_test,y_pred2)*100
                        ]})
```

## THE ALGHORITHM PREDICTION WAS TETSED 

```python
def predict_with_models(input_features):
    """
    Takes a list of features and returns predictions from both models.
    
    input_features: list or array of shape (1, n_features)
    Example: [[5.1, 3.5, 1.4, 0.2]]
    """
    rf_prediction = rf_model.predict(input_features)
    gb_prediction = gb_model.predict(input_features)

    return {
        "Random Forest Prediction": rf_prediction[0],
        "Gradient Boosting Prediction": gb_prediction[0]
    }
```

```python
# One test input (must match the number of features your models were trained on)
test_input = [[
    0.03, 0.85, 0.16, -0.80, 0.70, -0.57,
    0.93, -0.06, -0.35, -0.26, 0.50, 0.95,
    0.47, 0.19, -1.13, 0.19, -0.83, -0.18,
    0.35, 0.01, -0.23, -0.49, 0.00, -0.40,
    -0.45, 0.14, 0.25, 0.08, -0.34
]]  # Replace with real feature values

predictions = predict_with_models(test_input)

print(predictions)

```
